```python
import os
import pandas as pd

BASE_DIR   = "/kaggle/input/datasets/kmader/skin-cancer-mnist-ham10000"
CSV_PATH   = f"{BASE_DIR}/HAM10000_metadata.csv"
IMAGE_DIRS = [
    f"{BASE_DIR}/HAM10000_images_part_1",
    f"{BASE_DIR}/HAM10000_images_part_2",
]

# 1. Load metadata CSV
df = pd.read_csv(CSV_PATH)
print("Columns:", df.columns.tolist())
print()

# 2. Total number of images
print(f"Total images (from CSV): {len(df)}")
print()

# 3. Available classes (label column is 'dx')
class_counts = df["dx"].value_counts()
print(f"Available classes ({len(class_counts)}): {class_counts.index.tolist()}")
print()

# 4. Number of images per class
print("Images per class:")
print(class_counts.to_string())
print()

# 5. Cross-check with actual files on disk
image_files = set()
for d in IMAGE_DIRS:
    image_files.update(f for f in os.listdir(d) if f.lower().endswith((".jpg", ".jpeg", ".png")))
print(f"Image files found on disk: {len(image_files)}")
```

```python
import os
import pandas as pd

BASE_DIR  = "/kaggle/input/datasets/jakeshbohaju/brain-tumor"
CSV_PATH  = f"{BASE_DIR}/Brain Tumor.csv"
IMAGE_DIR = f"{BASE_DIR}/Brain Tumor/Brain Tumor"

# 1. Load metadata CSV
df = pd.read_csv(CSV_PATH)
print("Columns:", df.columns.tolist())
print()

# 2. Total number of images
print(f"Total images (from CSV): {len(df)}")
print()

# 3. Available classes (label column is 'Class': 0 = no tumor, 1 = tumor)
class_counts = df["Class"].value_counts()
print(f"Available classes ({len(class_counts)}): {class_counts.index.tolist()}")
print()

# 4. Number of images per class
print("Images per class:")
print(class_counts.to_string())
print()

# 5. Cross-check with actual files on disk
image_files = [f for f in os.listdir(IMAGE_DIR) if f.lower().endswith((".jpg", ".jpeg", ".png"))]
print(f"Image files found on disk: {len(image_files)}")
```
